```python
df['review_rating'] = pd.to_numeric(df['review_rating'], errors='coerce')
```

```python
df['review_rating']= df.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))

# This piece of code simply groups the ratings based on their categories and then the medium value is imputted in them according to the medium value of that category.
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()          # This changes the column names into small cases 
df.columns = df.columns.str.replace(' ','_') # This puts the underscore in place of space in between column names .
                                             # Overall this piece of code changes the columns names to snake case.
```

```python
df.info()   # Cheking the columns name for assuring that they are consistent .
```

```python
# Since the purchase_amount_(usd) is a bit odd so we will remove the usd from the column name.

df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.info()
```

```python
# Now the columns consistency is achieved .
# Now I will be doing some feature engineering .(Creating, modifying, or selecting useful columns (features) from raw data to help your model or analysis produce better results.)
```

```python
# Create a column age_group to differentiate the ages into categorical form for the easiness.
# The group is divide into 4 groups classifying the age groups of various people.
labels = ['Young Adult', 'Adult','Middle Aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
# Now creating a column named  purchase_frequency_days

frequency_mapping =
    'Fortnightly': 14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90   
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
